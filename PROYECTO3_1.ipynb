CENTRO DE AMPLITUD DE CADA COLUMNA

```python
#CREAR UNA FUNCIÓN QUE OBTENGA EL CENTRO AMPLITUD DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

def Amplitud (columns,data):
  amplitud = []
  for x in columns:
    c_a = (min(data[x]) + max(data[x])) /2
    amplitud.append(round(c_a,2))
  
  #print("Lista de medias: ", amplitud)

  #Crear un Data Frame con la metrica obtenida
  amp_df = pd.DataFrame()
  amp_df["Variable"] = columns
  amp_df["Centro de Amplitud"] = amplitud

  print('-----------------------------------')
  print('CENTRO DE AMPLITUD de cada variable')
  print('-----------------------------------')
  print(amp_df)
  pass
  
#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
Amplitud(columnas,data_nutricion)
```

MEDIA GEOMÉTRICA 

```python
#CREAR UNA FUNCIÓN QUE OBTENGA LA MEDIA GEOMETRICA DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

import numpy as np 

def MEDIA_GEOM(columns,data):
  media_geom = []
  for x in columns:
    a = np.log(data[x])
    media_geom.append(round(np.exp(a.sum()/len(a)),2))
  
  #Crear un Data Frame con la metrica obtenida
  mediageom_df = pd.DataFrame()
  mediageom_df["Variable"] = columns
  mediageom_df["Media Geométrica"] = media_geom

  print('----------------------------------')
  print('MEDIA GEOMETRICA de cada variable')
  print('----------------------------------')
  print(mediageom_df)

  pass


#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
MEDIA_GEOM(columnas,data_nutricion)
```

MEDIA ARMONICA

```python
#CREAR UNA FUNCIÓN QUE OBTENGA LA MEDIA ARMONICA DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

import numpy as np
import statistics as stats # estadística

def MEDIA_ARMON(columns,data):
  z = []
  for x in columns:
    media_arm = stats.harmonic_mean(data[x])
    z.append(round(media_arm,2))
  
  #Crear un Data Frame con la metrica obtenida
  df = pd.DataFrame()
  df["Variable"] = columns
  df["Media Armónica"] = z

  print('----------------------------------')
  print('MEDIA ARMÓNICA de cada variable')
  print('----------------------------------')
  print(df)
  pass

#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
MEDIA_ARMON(columnas,data_nutricion)
```

MEDIA RECORTADA 

AL 20% EN CADA LADO

```python
#CREAR UNA FUNCIÓN QUE OBTENGA LA MEDIA RECORTADA DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

import numpy as np
from scipy import stats as nstats

def MEDIA_RECORTADA(columns,data):
  z = []
  for x in columns:
    media_recortada = nstats.trim_mean(data[x], proportiontocut=0.20) #proportiontocut=0.20 Es la cantidad a recortar de cada extremo.
    z.append(round(media_recortada,2))
  
  #Crear un Data Frame con la metrica obtenida
  df = pd.DataFrame()
  df["Variable"] = columns
  df["Media Recortada"] = z

  print('----------------------------------')
  print('MEDIA RECORTADA de cada variable')
  print('----------------------------------')
  print(df)
  pass

#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
MEDIA_RECORTADA(columnas,data_nutricion)
```

```python
#CREAR UNA FUNCIÓN QUE OBTENGA EL RANGO INTERCUARTIL DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

import numpy as np


def INTER(columns,data):
  z = []
  for x in columns:
    
    #Obtiene los cuartiles 1 y 3 
    Q3,Q1 = np.percentile(data[x],[75, 25]) 
    
    #print("Q1: ", Q1)
    #print("Q3: ",Q3)

    #Obitiene el rango intercuartil
    r = Q3 - Q1 
    
    #print("Rango Intercuartil: ",r)
    z.append(round(r,2))
  
  #Crear un Data Frame con la metrica obtenida
  
  df = pd.DataFrame()
  df["Variable"] = columns
  df["RANGO INTERCUARTIL"] = z

  print('-----------------------------------')
  print('RANGO INTERCUARTIL de cada variable')
  print('-----------------------------------')
  print(df)

  pass


#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
INTER(columnas,data_nutricion)
```

```python
#CREAR UNA FUNCIÓN QUE OBTENGA LA VARIANZA DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

def VARIANZA (columns,data):
  z = []
  for x in columns:
    varianza = np.var(data[x])
    z.append(round(varianza,2))
  
  #Crear un Data Frame con la metrica obtenida
  
  df = pd.DataFrame()
  df["Variable"] = columns
  df["VARIANZA"] = z

  print('-----------------------------------')
  print('VARIANZA de cada variable')
  print('-----------------------------------')
  print(df)
  pass

#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
VARIANZA(columnas,data_nutricion)
```

DESVIACION ESTANDAR 

```python
#CREAR UNA FUNCIÓN QUE OBTENGA LA DESVIACION ESTÁNDAR DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

import statistics as stats

def DESVSTD(columns,data):
  z = []
  for x in columns:
    desv_estandar = stats.stdev(data[x])
    z.append(round(desv_estandar,2))
  
  #Crear un Data Frame con la metrica obtenida
  
  df = pd.DataFrame()
  df["Variable"] = columns
  df["DESV_STD"] = z

  print('-----------------------------------')
  print('DESV_STD de cada variable')
  print('-----------------------------------')
  print(df)
  
  pass

#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
DESVSTD(columnas,data_nutricion)
```

COEFICIENTE DE VARIACION 

```python
#CREAR UNA FUNCIÓN QUE OBTENGA LA DESVIACION ESTÁNDAR DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

def COEFVAR(columns,data):
  z = []
  for x in columns:
    media = data[x].mean()
    d_estandar = data[x].std()
    Coef_variacion = d_estandar/media*100
    z.append(round(Coef_variacion,2))
  
  #Crear un Data Frame con la metrica obtenida
  
  df = pd.DataFrame()
  df["Variable"] = columns
  df["COEFICIENTE_VARIACION "] = z

  print('-----------------------------------')
  print('COEFICIENTE VARIACION de cada variable')
  print('-----------------------------------')
  print(df)

  pass

#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
COEFVAR(columnas,data_nutricion)
```

PERCENTILES 

```python
#CREAR UNA FUNCIÓN QUE OBTENGA EL PERCENTIL DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#LA FUNCION PREGUNTA AL USUARIO EL PERCENTIL QUE DESEA OBTENER DE LOS DATOS
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

def PERCENTIL(columns,data):
  p = float(input("Indique el percetil a obtner (decimal): "))
  #print(p)
  z = []
  for x in columns:
    Q = np.percentile(a = data[x], q = p) 
    #print("Percentil Q: ",Q)
    z.append(round(Q,2))
  
  #Crear un Data Frame con la metrica obtenida
  
  df = pd.DataFrame()
  df["Variable"] = columns
  df[p] = z

  print('-----------------------------------')
  print('PERCENTIL de cada variable')
  print('-----------------------------------')
  print(df)
  pass

#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
PERCENTIL(columnas,data_nutricion)
```

CURTOSIS

```python
#CREAR UNA FUNCIÓN QUE OBTENGA LA CURTOSIS DE CADA UNA 
#DE LAS COLUMNAS DEL DATA FRAME
#CREAR UN DATA FRAME CON LOS RESULTADOS OBTENIDOS 

from scipy.stats import kurtosis

def CURTOSIS(columns,data):
  z = []
  y = []

  for x in columns:
    curtosis =  kurtosis(data[x], fisher=True)
    if curtosis >0:
      tipo = "Leptocúrtica"
    elif curtosis ==0:
      tipo = "Mesocúrtica"
    else:
      tipo = "Platicúrtica"
    z.append(round(curtosis,2))
    y.append(tipo)
  
  #Crear un Data Frame con la metrica obtenida
  
  df = pd.DataFrame()
  df["Variable"] = columns
  df["TIPO"] = y
  df["CURTOSIS"] = z

  print('-----------------------------------')
  print('CURTOSIS de cada variable')
  print('-----------------------------------')
  print(df)
  pass

#Obtener el array de las columnas de la DATA 
columnas = data_nutricion.columns.values
#print(columnas)
CURTOSIS(columnas,data_nutricion)
```
